```python
# Example of GeoCoding
import geopy
#dir(geopy)
geo_data_frame = pandas.read_csv("http://pythonhow.com/supermarkets.csv")
geo_data_frame
```

```python
from geopy.geocoders import Nominatim

nominatim = Nominatim()
address = nominatim.geocode("3995 23rd St, San Francisco, CA 94114")
lat = address.latitude
long = address.longitude

print(lat)
print(long)
```

```python
import pandas

address_df = pandas.read_csv("supermarkets.csv")
address_df
```

```python
address_df["Address"] = address_df["Address"] + ", " + address_df["City"] + ", " + address_df["State"] + ", " + address_df["Country"]
address_df
```

```python
address_df["Coordinates"] = address_df["Address"].apply(nominatim.geocode)
address_df
```

```python
address_df.Coordinates[0]
```

```python
address_df.Coordinates[0].latitude
```

```python
address_df["Latitude"] = address_df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
address_df["Longitude"] = address_df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
address_df
```
